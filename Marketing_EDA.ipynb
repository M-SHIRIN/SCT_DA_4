```python
# Cell 1: Uploading the file safely
import io
from google.colab import files

uploaded = files.upload()

# Automatically fetch the name of whatever file you just uploaded
uploaded_file_name = list(uploaded.keys())[0]
print(f"Successfully uploaded file: {uploaded_file_name}")
```

```python
# Cell 2: Library imports, custom configurations, and dynamic file loading
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Apply the custom dual-tone styling palette
sns.set_theme(style="whitegrid")
custom_palette = ["#0f766e", "#f43f5e"]
sns.set_palette(custom_palette)

# Dynamically load the exact file name uploaded in Step 1
df = pd.read_csv(uploaded_file_name)
print(f"Dataset loaded successfully! Total records: {len(df)}")
df.head()
```

```python
# Cell 3: Target variable distribution
plt.figure(figsize=(6, 4))
sns.countplot(x='deposit', data=df)
plt.title('Deposit Subscription Distribution', fontsize=12, pad=12, weight='bold', color='#0f766e')
plt.show()
```

```python
# Cell 4: Horizontal distribution breakdown by occupation
plt.figure(figsize=(12, 6))
sns.countplot(y='job', hue='deposit', data=df)
plt.title('Deposit Subscription by Job', fontsize=12, pad=12, weight='bold', color='#0f766e')
plt.show()
```

```python
# Cell 5: Grouped bar chart by education
plt.figure(figsize=(7, 4.5))
sns.countplot(x='education', hue='deposit', data=df)
plt.title('Deposit by Education Level', fontsize=12, pad=12, weight='bold', color='#0f766e')
plt.show()
```

```python
# Cell 6: Histogram with KDE trendline overlay
plt.figure(figsize=(6, 4))
sns.histplot(df['age'], bins=20, kde=True, color='#0f766e')
plt.title('Age Distribution', fontsize=12, pad=12, weight='bold', color='#0f766e')
plt.show()
```

```python
# Cell 7: Box plot showing outlier clusters
plt.figure(figsize=(6, 4))
sns.boxplot(x='deposit', y='balance', data=df)
plt.title('Balance vs Deposit', fontsize=12, pad=12, weight='bold', color='#0f766e')
plt.ylim(-1000, 10000) # Prevents massive outlier balances from squishing the box lines
plt.show()
```

```python
# Cell 8: Comparative outreach channel performance
plt.figure(figsize=(6, 4))
sns.countplot(x='contact', hue='deposit', data=df)
plt.title('Contact Method vs Deposit', fontsize=12, pad=12, weight='bold', color='#0f766e')
plt.show()
```
